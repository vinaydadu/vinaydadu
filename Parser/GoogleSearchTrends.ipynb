```python
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['Ambani'])
# Interest by Region
df = pytrend.interest_by_region()
df.head(200)
```

```python
df.reset_index().plot(x='geoName', y='Ambani', figsize=(120, 10), kind ='bar')
```

```python
df = pytrend.trending_searches(pn='india')
df.head()
```

```python
df = pytrend.today_searches(pn='US')
df
```

```python
# Get Google Top Charts
df = pytrend.top_charts(2020, hl='en-US', tz=300, geo='GLOBAL')
df.head()
```

```python
# Get Google Keyword Suggestions
keywords = pytrend.suggestions(keyword='sputnik v')
df = pd.DataFrame(keywords)
df.drop(columns= 'mid')   # This column makes no sense
```

```python
# Related Queries, returns a dictionary of dataframes
related_queries = pytrend.related_queries()
related_queries.values()
```

```python
# Related Topics, returns a dictionary of dataframes
related_topic = pytrend.related_topics()
related_topic.values()
```
